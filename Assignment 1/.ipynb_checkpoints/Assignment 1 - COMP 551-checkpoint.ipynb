```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from IPython.display import display, HTML
import random

########### Question 1 ###########
## 1

# data
df_train = pd.read_csv("hwk1_datasets/Datasets/Dataset_1_train.csv", header=None)
df_valid = pd.read_csv("hwk1_datasets/Datasets/Dataset_1_valid.csv", header=None)
df_test = pd.read_csv("hwk1_datasets/Datasets/Dataset_1_test.csv", header=None)

# training
data_train = df_train.values
x_train = np.array(data_train[:,0])
x_train = np.vstack(x_train)
y_train = np.array(data_train[:,1])
y_train = np.vstack(y_train)

# validation
data_valid = df_valid.values
x_valid = np.array(data_valid[:,0])
x_valid = np.vstack(x_valid)
y_valid = np.array(data_valid[:,1])
y_valid = np.vstack(y_valid)

# testing
data_test = df_test.values
x_test = np.array(data_test[:,0])
x_test = np.vstack(x_test)
y_test = np.array(data_test[:,1])
y_test = np.vstack(y_test)

plt.plot(x_train,y_train,'o')
plt.title("Dataset 1")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# manipulate to m+1 columns with m=20
X_train = np.hstack((x_train, np.ones([x_train.size,1])))
X_valid = np.hstack((x_valid, np.ones([x_valid.size,1])))
X_test = np.hstack((x_test, np.ones([x_test.size,1])))


for i in range(19):
    x_tmp = np.multiply(X_train[:,0],X_train[:,i])
    x_tmp = np.vstack(x_tmp)
    X_train = np.hstack((x_tmp,X_train))
    
    x_tmp = np.multiply(X_valid[:,0],X_valid[:,i])
    x_tmp = np.vstack(x_tmp)
    X_valid = np.hstack((x_tmp,X_valid))

    x_tmp = np.multiply(X_test[:,0],X_test[:,i])
    x_tmp = np.vstack(x_tmp)
    X_test = np.hstack((x_tmp,X_test))
```

```python
# closed-form
w = np.matmul(np.matmul(np.linalg.inv(np.matmul(X_train.T,X_train)),X_train.T),y_train)
```

```python
# MSE function
def MSE(w,x,y):
    return (np.matmul(np.transpose(np.subtract(y,np.matmul(x,w))),np.subtract(y,np.matmul(x,w)))) / y.shape[0]
    


# training data
print("Training Error w/out L2 Regularization: ", np.asscalar(MSE(w,X_train,y_train)))

# validation data
print("Validation error w/out L2 Regularization: ", np.asscalar(MSE(w,X_valid,y_valid)))

# Quality of the fit
# Without regularization, the fit seems to be overfitted. This is apparent from the sharp "ups and downs" of the line. 
# Especially at around -1 and 1, the fit seems to show trends that are no realistically represented in the data.
```

```python
## 2
# L2 regularization function
def L2(x,y,L):
    I = np.identity(np.matmul(x.T,x).shape[0])
    return np.matmul(np.matmul(np.linalg.inv(np.add(np.matmul(x.T,x),I*L)),x.T),y)


# training data
L = np.arange(0,1.001,0.001) # lambda from 0 to 1 

# initialize training and validation error arrays for storage
MSE_train = np.empty(shape=(0))
MSE_valid = np.empty(shape=(0))

for i in range(L.size):
    w_L2 = L2(X_train,y_train,L[i])

    MSE_train = np.append(MSE_train, MSE(w_L2,X_train,y_train))
    # print("Training: ", MSE_train)
    # graph(lambda xs: xs**20*w_L2[0] + xs**19*w_L2[1] + xs**18*w_L2[2] + xs**17*w_L2[3] + xs**16*w_L2[4] + xs**15*w_L2[5] + xs**14*w_L2[6] + xs**13*w_L2[7] + xs**12*w_L2[8] + xs**11*w_L2[9] + xs**10*w_L2[10] + xs**9*w_L2[11] + xs**8*w_L2[12] + xs**7*w_L2[13] + xs**6*w_L2[14] + xs**5*w_L2[15] + xs**4*w_L2[16] + xs**3*w_L2[17] + xs**2*w_L2[18] + xs*w_L2[19] + w_L2[20], np.arange(-1,1,0.1), x_train, y_train)

    MSE_valid = np.append(MSE_valid, MSE(w_L2,X_valid,y_valid))
    # print("Validation: ", MSE_valid)
    # graph(lambda xs: xs**20*w_L2[0] + xs**19*w_L2[1] + xs**18*w_L2[2] + xs**17*w_L2[3] + xs**16*w_L2[4] + xs**15*w_L2[5] + xs**14*w_L2[6] + xs**13*w_L2[7] + xs**12*w_L2[8] + xs**11*w_L2[9] + xs**10*w_L2[10] + xs**9*w_L2[11] + xs**8*w_L2[12] + xs**7*w_L2[13] + xs**6*w_L2[14] + xs**5*w_L2[15] + xs**4*w_L2[16] + xs**3*w_L2[17] + xs**2*w_L2[18] + xs*w_L2[19] + w_L2[20], np.arange(-1,1,0.1), x_valid, y_valid)


MSE_train = np.vstack(MSE_train)
MSE_valid = np.vstack(MSE_valid)
#MSE_total = np.hstack((MSE_train,MSE_valid))
L = np.vstack(L)
#MSE_total = np.hstack((L,MSE_total))
#print(MSE_total)
#print(np.amin(MSE_valid))

# plot testing and validation MSEs
plt.plot(L[1:,],MSE_train[1:,],label="Training")
plt.plot(L[1:,],MSE_valid[1:,],label="Validation")
plt.title("MSEs of Training and Validation Data")
plt.xlabel("Lambda")
plt.ylabel("MSE")
plt.legend()
plt.show()
```

```python
for i in range(L.size):
    if np.equal(np.around(np.amin(MSE_valid),decimals=5),np.around(MSE_valid[i],decimals=5)):
        L_opt = L[i]

        

w_opt = L2(X_train,y_train,L_opt)

# training and testing data
graph(lambda xs: xs**20*w_opt[0] + xs**19*w_opt[1] + xs**18*w_opt[2] + xs**17*w_opt[3] + xs**16*w_opt[4] + xs**15*w_opt[5] + xs**14*w_opt[6] + xs**13*w_opt[7] + xs**12*w_opt[8] + xs**11*w_opt[9] + xs**10*w_opt[10] + xs**9*w_opt[11] + xs**8*w_opt[12] + xs**7*w_opt[13] + xs**6*w_opt[14] + xs**5*w_opt[15] + xs**4*w_opt[16] + xs**3*w_opt[17] + xs**2*w_opt[18] + xs*w_opt[19] + w_opt[20], np.arange(-1.1,1.1,0.1), x_train, y_train, "Training w/L2 Regularization")
graph(lambda xs: xs**20*w_opt[0] + xs**19*w_opt[1] + xs**18*w_opt[2] + xs**17*w_opt[3] + xs**16*w_opt[4] + xs**15*w_opt[5] + xs**14*w_opt[6] + xs**13*w_opt[7] + xs**12*w_opt[8] + xs**11*w_opt[9] + xs**10*w_opt[10] + xs**9*w_opt[11] + xs**8*w_opt[12] + xs**7*w_opt[13] + xs**6*w_opt[14] + xs**5*w_opt[15] + xs**4*w_opt[16] + xs**3*w_opt[17] + xs**2*w_opt[18] + xs*w_opt[19] + w_opt[20], np.arange(-1.1,1.1,0.1), x_test, y_test, "Testing w/L2 Regularization")
print("Optimal Lambda: ", np.asscalar(L_opt))
print("Testing MSE: ", np.asscalar(MSE(w_opt,X_test,y_test)))


# The fit with L2 regularization better represents the dataset. The previously seen overfitting is no longer observed, and
# the curve has smoothed out compared to without regularization. However, at -1 and below, the model seems to not reflect
# the test data properly. This is likely due to the fit being generated from the training data which shows this pattern, 
# as well as have no data at less than -1.
```

```python
########### Question 2 ###########
df_train = pd.read_csv("hwk1_datasets/Datasets/Dataset_2_train.csv", header=None)
df_valid = pd.read_csv("hwk1_datasets/Datasets/Dataset_2_valid.csv", header=None)
df_test = pd.read_csv("hwk1_datasets/Datasets/Dataset_2_test.csv", header=None)

# training
data_train = df_train.values
x_train = np.array(data_train[:,0])
x_train = np.vstack(x_train)
y_train = np.array(data_train[:,1])
y_train = np.vstack(y_train)

# validation
data_valid = df_valid.values
x_valid = np.array(data_valid[:,0])
x_valid = np.vstack(x_valid)
y_valid = np.array(data_valid[:,1])
y_valid = np.vstack(y_valid)

# testing
data_test = df_test.values
x_test = np.array(data_test[:,0])
x_test = np.vstack(x_test)
y_test = np.array(data_test[:,1])
y_test = np.vstack(y_test)

plt.plot(x_train,y_train,'o')
plt.title("Dataset 2 Visualization")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
## 1

# manipulate to m+1 columns with m=20
X_train = np.hstack((x_train, np.ones([x_train.size,1])))
X_valid = np.hstack((x_valid, np.ones([x_valid.size,1])))
X_test = np.hstack((x_test, np.ones([x_test.size,1])))

# shuffle data for SGD
def shuffle(x):
    np.random.permutation(x)
```

```python
# initialize 2 random weights
w1 = np.random.random((2,1))
w2 = np.random.random((2,1))

train = []
valid = []
train.append(MSE(w2,X_train,y_train))
valid.append(MSE(w2,X_valid,y_valid))

epoch = 1

# stochastic gradient descent
alpha = 0.000001    # step size
e = 0.005
shuffle(X_train)

while(True):
    for i in range(X_train.shape[0]):
        X = X_train[i]
        Y = np.asscalar(y_train[i])
        X = X[np.newaxis,:]

        w1t = w1
        w1 = w2
        w2 = w1t - alpha*2*(np.subtract(np.matmul(np.matmul(X.T,X),w1t),(X.T*Y)))
    
    shuffle(X_train)
        
    # find MSE on training and validation set
    train.append(MSE(w2,X_train,y_train))
    valid.append(MSE(w2,X_valid,y_valid))
    epoch += 1
    
    if epoch >= 15000 or np.all(np.absolute(np.subtract(w2,w1))<e):  # stop if number of epochs get too high or below epsilon
        break
    

MSE_train = np.asarray(train, dtype=np.float32)
MSE_train = np.squeeze(MSE_train)
MSE_valid = np.asarray(valid, dtype=np.float32)
MSE_valid = np.squeeze(MSE_valid)

plt.subplot(121)   
plt.plot(np.arange(0,epoch,1),MSE_train,linestyle = '-.', color = 'k', label="Training", zorder=2)
plt.plot(np.arange(0,epoch,1),MSE_valid, color = 'r', label="Validation",zorder=1)
plt.legend()
plt.title("MSEs of Training and Validation")
plt.xlabel("Epoch")
plt.ylabel("MSE")

plt.subplot(122)
graph(lambda xs: xs*w2[0] + w2[1], np.arange(0,1.5,0.1), x_train, y_train, "SGD Training")
```

```python
## 2
start = 0.0005
step = 0.0002
end = 0.005
alpha_list = np.arange(start,end, step)    # alpha range 
e = 0.005

min_MSE = []

# initialize 2 random weights
w1 = np.random.random((2,1))
w2 = np.random.random((2,1))

for i in range(alpha_list.size):    
    valid = []
    
    # find MSE on training and validation set
    valid.append(MSE(w2,X_valid,y_valid))
    epoch = 1
    
    while(True):
        for i in range(X_train.shape[0]):
            X = X_train[i]
            Y = np.asscalar(y_train[i])
            X = X[np.newaxis,:]

            w1t = w1
            w1 = w2
            w2 = w1t - alpha*2*(np.subtract(np.matmul(np.matmul(X.T,X),w1t),(X.T*Y)))

        shuffle(X_train)

        epoch += 1
        
        if epoch >= 5000 or np.all(np.absolute(np.subtract(w2,w1))<e):  # stop if number of epochs get too high or below epsilon
            break
    
    # find minimum MSE of validation set 
    MSE_valid = np.asarray(valid, dtype=np.float32)
    MSE_valid = np.squeeze(MSE_valid)
    min_MSE.append(np.amin(MSE_valid))


alpha_index = min_MSE.index(min(min_MSE))
alpha_opt = start + step*alpha_index
print("Best alpha: ", alpha_opt)
```

```python
# MSE on test data using weights determined from optimum alpha

# find weights from alpha_opt on training data

# generate random weights
w1 = np.random.random((2,1))
w2 = np.random.random((2,1))
epoch = 1

# gradient descent
e = 0.005

while(True):
    #print(epoch)
    for i in range(X_train.shape[0]):
        X = X_train[i]
        Y = np.asscalar(y_train[i])
        X = X[np.newaxis,:]

        w1t = w1
        w1 = w2
        w2 = w1t - alpha_opt*2*(np.subtract(np.matmul(np.matmul(X.T,X),w1t),(X.T*Y)))

    shuffle(X_train)

    epoch += 1

    if epoch >=15000 or np.all(np.absolute(np.subtract(w2,w1))<e):  # stop if number of epochs get too high or below epsilon
        break

print("Testing Error with Weights from Optimum Alpha: ", np.asscalar(MSE(w2,X_test,y_test)))
graph(lambda xs: xs*w2[0] + w2[1], np.arange(0,1.5,0.1), x_test, y_test, "Testing w/ Optimum Alpha")
```

```python
########### Question 3 ###########
## 1

f = open("hwk1_datasets/Datasets/communities.data","r")
file = f.read()
f.close()

parse = file.splitlines()
data = np.array([p.split(",") for p in parse])
data = np.delete(data,np.s_[0:5],axis=1)                            # remove non-predictive features
data[data == "?"] = "-10000"                                        # change all ? into -10000
data = data.astype(float)                                           # cast data into float 
```

```python
# fill in missing data
for i in range(data.shape[1]):
    total = 0
    count = 0
    
    for j in range(data.shape[0]):
        if data[j][i] == -10000:
            continue
            
        total += data[j][i]
        count += 1
        
    mean = total / count
    
    for k in range(data.shape[0]):
        if data[k][i] == -10000:
            data[k][i] = mean

# save completed dataset
name = "hwk1_datasets/Datasets/Q3_completed_dataset.csv"
np.savetxt(name,data,delimiter=",")


# Is this a good choice?
# This is likely a safe choice when the number of missing datapoints are not too high. If most datapoints are present (for
# each feature), then the mean represents the majority of the feature data. However, if too many datapoints for a feature
# are missing, then the mean may not be a good representation of the feature. 

# What else might you use?
# Another method may be to find the distribution of the data of the feature of interest, and sampling the missing datapoints
# from this distribution.
```

```python
## 2

# five sections
r = int((data.shape[0]+1)/5)

one = data[0:r,:]
two = data[r:2*r,:]
three = data[2*r:3*r,:]
four = data[3*r:4*r,:]
five = data[4*r:,:]


# save 4 80-20 splits into csv
for i in range(5):
    data_list = [one, two, three, four, five]  # place into list
    num = str(i+1)
    
    name = "hwk1_datasets/Datasets/8020/CandC-test%s.csv" %num
    np.savetxt(name,data_list[i],delimiter=",")
    del data_list[i]

    data_rem = np.array(data_list)
    data_rem = [item for sublist in data_rem for item in sublist]

    name = "hwk1_datasets/Datasets/8020/CandC-train%s.csv" %num
    np.savetxt(name,data_rem,delimiter=",")
```

```python
# linear regression w/out L2
MSE_list = []      # store MSE 

for i in range(5):
    # load data
    num = str(i+1)
    name_train = "hwk1_datasets/Datasets/8020/CandC-train%s.csv" %num
    name_test = "hwk1_datasets/Datasets/8020/CandC-test%s.csv" %num
    df_train = pd.read_csv(name_train, header=None)
    df_test = pd. read_csv(name_test, header=None)

    # training
    data_train = df_train.values
    x_train = np.array(data_train[:,0:-1])
    X_train = np.ones([x_train.shape[0],1])
    X_train = np.hstack((x_train,X_train))
    y_train = np.array(data_train[:,-1])

    # testing
    data_test = df_test.values
    x_test = np.array(data_test[:,0:-1])
    X_test = np.ones([x_test.shape[0],1])
    X_test = np.hstack((x_test,X_test))
    y_test = np.array(data_test[:,-1])

    # closed-form
    w = np.matmul(np.matmul(np.linalg.inv(np.matmul(X_train.T,X_train)),X_train.T),y_train)
    MSE_list.append(MSE(w,X_test,y_test))

    # save weights
    name = ("Parameters_Learned/Q3_no_L2_Set%s.txt")
    np.savetxt(name %str(i+1),w, delimiter=",")
    
MSE_mean = sum(MSE_list) / len(MSE_list)
print("MSE without L2 regularization: ", MSE_mean)
```

```python
## 3 
# L2 linear regression

# training data
start = 0
step = 0.1
end = 5.1
L = np.arange(start,end,step) # lambda from 0 to 5

# initialize training and validation error arrays for storage
MSE_mean_list = []
w_list = []
for i in range(L.size):
    MSE_list = []
    for j in range(5):
        # load data
        num = str(j+1)
        name_train = "hwk1_datasets/Datasets/8020/CandC-train%s.csv" %num
        name_test = "hwk1_datasets/Datasets/8020/CandC-test%s.csv" %num
        df_train = pd.read_csv(name_train, header=None)
        df_test = pd. read_csv(name_test, header=None)

        # training
        data_train = df_train.values
        x_train = np.array(data_train[:,0:-1])
        X_train = np.ones([x_train.shape[0],1])
        X_train = np.hstack((x_train,X_train))
        y_train = np.array(data_train[:,-1])

        # testing
        data_test = df_test.values
        x_test = np.array(data_test[:,0:-1])
        X_test = np.ones([x_test.shape[0],1])
        X_test = np.hstack((x_test,X_test))
        y_test = np.array(data_test[:,-1])
        
        w_L2 = L2(X_train,y_train,L[i])
        MSE_list.append(MSE(w_L2,X_test,y_test))
    
        # add to weight list
        w_list.append(w_L2)
    
    MSE_mean = sum(MSE_list) / len(MSE_list)
    MSE_mean_list.append(MSE_mean)
    
    # save weights
    name = ("Parameters_Learned/Q3_L2_lambda_%s.txt")
    np.savetxt(name %str(L[i]),np.array(w_list),delimiter=",")


lambda_index = MSE_mean_list.index(min(MSE_mean_list))
lambda_opt = start + step*lambda_index
print("Best lambda: ", lambda_opt)

MSE_mean_list = np.vstack((np.arange(start,end,step),MSE_mean_list))
name = ("Lambda_MSE_list.csv")
np.savetxt(name,MSE_mean_list,delimiter=",")
```
